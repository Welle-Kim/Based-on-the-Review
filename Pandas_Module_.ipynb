```python
## iloc[] : 데이터의 행/열 인덱싱된 값을 기준으로 출력
df2.iloc[0]
df2.iloc[1:5]
df2.iloc[1:5, 0]
df2.iloc[1:5, 1:]
df2.iloc[1:5, [0,1,3]]
```

```python
lst=[]
for i in df2.index:
  lst.append(int(df2.loc[i,["휘발유"]]))

print(lst)
```

```python
df2["휘발유"] = lst
df2.dtypes
```

```python
df3=df1.loc[1:10, ["셀프여부","휘발유", "경유", "상표"]]
print(df3.dtypes)
print(len(df3))
```
