```python
import pandas as pd
import numpy as np
```

```python
# pandas.Series([값])
ser=pd.Series([1, 3, 5, np.nan, 6, 9])
print(ser)
print(ser[:2])
print(type(ser))
```

```python
# DataFrmae 생성
# pandas.date_range("시작일", periods=숫자) : 시작일부터 입력한 숫자만큼 날짜를 발생시키는 함수

date_lst=pd.date_range("20220301", periods=6)
print(date_lst)
```

```python
# numpy.random.randn(6,4)  # randn(행, 열) 행개수/열개수 정의
data=np.random.randn(6,4)
print(data)
```

```python
### DataFrame 생성(배열(리스트))
df1=pd.DataFrame(data)
display(df1)
```

```python
df1=pd.DataFrame(data, columns=["A", "B", "C", "D"])
display(df1)
```

```python
df1=pd.DataFrame(data, columns=["A", "B", "C", "D"], index=date_lst)
display(df1)
```

```python
# df1을 df2로 복사
# df2에 새로운 필드(열) 추가

df2 = df1.copy()
df3 = df1

display(df2)
display(df3)
```

```python
df2["E"]=["aa", "bb", "aa", "cc", "bb", "ff"]
display(df1)
display(df2)
```

```python
df3["E"]=["aa", "bb", "aa", "cc", "bb", "ff"]
display(df1)
display(df3)
```

```python
del df1["E"]   # del 객체 삭제

display(df1)
display(df3)
display(df2)
```

```python
df2[df2["E"].isin(["cc", "ff"])] # E열에 "cc", "ff"가 있는 값 모두 출력
```

```python
df2[df2["E"].isin(["cc", "ff"])][["C", "D"]]
```

```python
df2.loc[df2["E"] == "aa", ["C", "D"]]
```

```python
df2.loc[(df2["E"] == "aa") | (df2["E"] == "bb"), ["C", "D"]]
```
