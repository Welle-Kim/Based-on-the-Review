```python
# 여러 차트 그리기

import matplotlib.pyplot as plt

x =  [0,1,2,3,4,5]
y1 = [4,7,6,8,2,9]
y2 = [6,9,12,18,22,14]
y3 = [9,22,3,16,10,15]

plt.plot(x, y1, "r--", x, y2, "bs", x, y3, "g^")   # y값 차트 출력(비추천)

plt.show()
```

```python
#여러차트 그리기
import matplotlib.pyplot as plt

x =  [0,1,2,3,4,5]
y1 = [4,7,6,8,2,9]
y2 = [6,9,12,18,22,14]
y3 = [9,22,3,16,10,15]

plt.plot(x, y1, "--", x, y2, "-.", x, y3, ":")   # y값 차트 출력(비추천)

plt.show()
```

```python
# 여러 차트 그리기

import matplotlib.pyplot as plt

x =  [0,1,2,3,4,5]
y1 = [4,7,6,8,2,9]
y2 = [6,9,12,18,22,14]
y3 = [9,22,3,16,10,15]

plt.plot(x, y3, "r", label="y3-data")   
plt.plot(x, y1, "g", marker="^", label="y1-data")   

plt.legend()   # 범례 표시
plt.title("Graph Title", size=16)

plt.show()
```
